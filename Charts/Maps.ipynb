# Creating Maps



```python
import numpy as np
import pandas as pd
from PIL import Image
import folium
```

```python
df = pd.read_csv('Police_Department_Incidents_-_Previous_Year__2016_.csv')
df.head(), df.shape
```

```python
df = df[['PdDistrict', 'X', 'Y', 'Location', 'PdId']].copy()
```

```python
df.head()
```

```python
df.rename(columns={'PdDistrict':'Neighbourhood'}, inplace=True)
df.head()
```

```python
df_sanFran = df.groupby(['Neighbourhood']).size().reset_index(name='count')
df_sanFran.head()
```

```python
san_geo = r'SF_Neighborhoods.geojson'
san_fran_map = folium.Map(location=[37.77, -122.42], zoom_start=12)
```

```python

san_fran_map.choropleth(
    geo_data=san_geo,
    data=df_sanFran,
    columns=['Neighbourhood', 'count'],
    key_on = 'Feature.properties.DISTRICT',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crime Rate in San Francisco'
)

# display map
san_fran
```
